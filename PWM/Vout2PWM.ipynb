```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Vout2PWM(Vout):
    if Vout < 4.57:
        return 0
    elif Vout < 21.9:
        return (Vout - 0.31276671) / 0.30431419
    else:
        return 71
```

```python
# 生成输入数据
vout = np.arange(0, 22, 0.1)
# 计算输出数据
duty_cycle = [Vout2PWM(v) for v in vout]
```

```python
data = np.loadtxt('/Users/hyx020222/Documents/GitHub/NCSLab/PWM/data_PWM2Vout.txt')

x = data[:, 0]
z = data[:, 2]
```

```python
# 绘制曲线
plt.scatter(z, x)
plt.plot(vout, duty_cycle)

plt.xlabel('Vout (V)')
plt.ylabel('DUTY_CYCLE (%)')

plt.show()
```
