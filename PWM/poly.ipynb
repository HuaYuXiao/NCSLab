```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('/home/hyx020222/NCSLab/PWM/data_PWM2Vout.txt')

x = data[:, 0]
y = data[:, 1]
z = data[:, 2]

x_part = x[2:8]
z_part = z[2:8]
```

```python
coefficients_xy = np.polyfit(x, y, 1)
poly_xy = np.poly1d(coefficients_xy)

coefficients_xz = np.polyfit(x_part, z_part, 1)
poly_xz = np.poly1d(coefficients_xz)
```

```python
plt.scatter(x, y)
plt.plot(x, poly_xy(x))

plt.scatter(x, z)
plt.plot(x_part, poly_xz(x_part))

plt.show()
```
